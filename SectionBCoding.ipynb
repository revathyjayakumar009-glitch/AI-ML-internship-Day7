 **Q1: 
Write program to check palindrome** 

```python
Str = input("Enter the String")
if Str==Str[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

**Q2: 
Find largest number in list**

```python
List = []
n=int(input("How many numbers inserted into the list:"))
for i in range(n):
    num = int(input("Enter number: "))
    List.append(num)
print(List)

print("Largest number in a list:",max(List))

      
```

**Q3: 
Count vowels in string**

```python
Str = input("Enter the string:")
vowels="aeiouAEIOU"
count=0

for c in Str:
    if c in vowels:
        count+=1

print("Number of Vowels:",count)
```
